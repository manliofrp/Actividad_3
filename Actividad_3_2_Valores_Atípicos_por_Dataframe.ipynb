```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# 2020

```python
df_2020 = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2020', skiprows=5)
df_2020
```

```python
df_2020.info()
```

```python
# Filtro por columnas

numericas = df_2020.iloc[:,[8,9,10,11,12,13,14,18]]
strings = df_2020.iloc[:,[0,1,2,3,4,5,7,15,16,17]]
```

```python
numericas.isnull().sum()
```

```python
numericas2 = numericas.fillna(method='bfill')
numericas2
```

```python
numericas2.isnull().sum()
```

```python
strings.isnull().sum()
```

```python
strings = strings.fillna(method='ffill')
strings = strings.fillna("Desconocido")
strings
```

```python
strings.isnull().sum()
```

```python
fig = plt.figure(figsize=(20,10))
numericas2.plot(kind='box',vert=False)
plt.title('Valores atípicos del DataFrame')
plt.show() #Dibujamos el diagrama
```

```python
# Metodo Desviacion Estandar
y = numericas2
Limite_Superior = y.mean() + 3*y.std()
Limite_Inferior = y.mean() - 3*y.std()
print('Límite superior permitido', Limite_Superior)
print('Límite inferior permitido', Limite_Inferior)
```

```python
outliers = numericas2[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
num3 = numericas2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
num3
```

```python
valores_nulos=num3.isnull().sum()
valores_nulos
```

```python
data_clean2020 = num3.copy()
data_clean2020 = data_clean2020.fillna(round(num3.mean(),1))
data_clean2020
```

```python
valores_nulos=data_clean2020.isnull().sum()
valores_nulos
```

```python
Datos_limpios_2020 = pd.concat([strings, data_clean2020],axis=1)
Datos_limpios_2020
```

```python
Datos_limpios_2020.isnull().sum()
```

```python
#Convertir dataframe a csv
Datos_limpios_2020.to_csv("Datos_limpios_2020.csv")
```

```python
#Descargamos
from google.colab import files
files.download('Datos_limpios_2020.csv')
```

# 2021

```python
df_2021 = pd.read_excel('Gastos y costos 20-23.xlsx', sheet_name = '2021', skiprows=3)
df_2021
```

```python
df_2021.info()
```

```python
numericas_2021 = df_2021.iloc[:,[8,9,10,11,12,13,14]]
strings_2021 = df_2021.iloc[:,[0,1,2,3,4,5,6,7,15,16,17]]
```

```python
numericas_2021.isnull().sum()
```

```python
strings_2021.isnull().sum()
```

```python
strings_2021 = strings_2021.fillna(method='bfill')
strings_2021 = strings_2021.fillna("Desconocido")
strings_2021
```

```python
strings_2021.isnull().sum()
```

```python
fig = plt.figure(figsize=(20,10))
numericas_2021.plot(kind='box',vert=False)
plt.title('Valores atípicos del DataFrame')
plt.show() #Dibujamos el diagrama
```

```python
y = numericas_2021

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print('Limite superior permitido', Limite_Superior_iqr)
print('Limite superior permitido', Limite_Inferior_iqr)
```

```python
num_iqr_2021 = numericas_2021[(y<=Limite_Superior)&(y>=Limite_Inferior)]
num_iqr_2021
```

```python
valores_nulos = num_iqr_2021.isnull().sum()
valores_nulos
```

```python
num_iqr_2021_2 = num_iqr_2021.copy()
num_iqr_2021_2 = num_iqr_2021_2.fillna(round(num_iqr_2021.mean(),1)) #El 1 es porque está redondeando
num_iqr_2021_2
```

```python
valores_nulos = num_iqr_2021_2.isnull().sum()
valores_nulos
```

```python
Datos_limpios_2021 = pd.concat([strings_2021, num_iqr_2021_2],axis=1)
Datos_limpios_2021
```

```python
Datos_limpios_2021.isnull().sum()
```

```python
#Convertir dataframe a csv
Datos_limpios_2021.to_csv("Datos_limpios_2021.csv")
```

```python
#Descargamos
from google.colab import files
files.download('Datos_limpios_2021.csv')
```
